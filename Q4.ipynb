```python
from sklearn.datasets import make_moons, make_blobs
import matplotlib.pyplot as plt
import numpy as np
from Models import Perceptron, Adaline
```

```python
x = np.arange(-6.0, 6.0, 0.01)
y = np.power(x,3) - 12 * x + 2


noise = 10 * np.random.normal(size=x.size)

ydata = y + noise
# ydata = ydata.reshape(-1,1)
# x = x.reshape(-1,1)
print(x.shape, ydata.shape)
plt.scatter(x,ydata)
plt.show()
```

```python
adaline = Adaline(3)
adaline.train(np.array([np.array([i, np.square(i), np.power(i, 3)]) for i in x]), ydata, 10)
```

```python
print(x.shape, ydata.shape)
plt.scatter(x, ydata)
line_x = np.linspace(x.min() , x.max(), 100)

line_y = [-adaline._weights[1] * i - adaline._weights[2] * np.square(i) - adaline._weights[3] * np.power(i, 3) - adaline._weights[0] * adaline._bias  for i in line_x] 


print(adaline._bias, adaline._weights)

plt.plot(line_x, line_x, 'r-')
plt.ylim(ydata.min(), ydata.max())

plt.show()
```
